```python
from collections import deque
from math import inf

# Get problem information, store locations as tuples
# pacman = list(map(int, input().strip().split()))
# pacman = (pacman[0], pacman[1])

# food = list(map(int, input().strip().split()))
# food = (food[0], food[1])

# rows, cols = list(map(int, input().strip().split()))

# grid = [[x for x in input()] for _ in range(rows)]

# rows = 3
# cols = 9
# pacman = (1, 2)
# food = (1, 7)


# inputgrid = ['%%%%%%%%%','%-P----.%', '%%%%%%%%%']

pacman = (11, 9)
food = (2, 15)
rows = 13
cols = 20

inputgrid = ['%%%%%%%%%%%%%%%%%%%%',
'%----%--------%----%',
'%-%%-%-%%--%%-%.%%-%',
'%-%-----%--%-----%-%',
'%-%-%%-%%--%%-%%-%-%',
'%-----------%-%----%',
'%-%----%%%%%%-%--%-%',
'%-%----%----%-%--%-%',
'%-%----%-%%%%-%--%-%',
'%-%-----------%--%-%',
'%-%%-%-%%%%%%-%-%%-%',
'%----%---P----%----%',
'%%%%%%%%%%%%%%%%%%%%']

grid = [[char for char in line] for line in inputgrid]

print("Grid: ")
for line in grid:
    print("".join(line))

# Awesome way to make a graph
def make_graph(grid, rows, cols):
    graph = dict()
    
    moves = [(0, -1), (-1, 0), (1, 0), (0, 1)]
    valid = ['-', 'P', '.']
    
    for i in range(rows):
        for j in range(cols):
            
            if grid[i][j] not in valid:
                continue
                
            loc = (i, j)
            
            # initialize location neighbors to empty list
            graph[loc] = []
            
            # Check UP, DOWN, LEFT, RIGHT
            for move in moves:
                if i > 0 and i < rows and j > 0 and j < cols:
                    check = (i + move[0], j + move[1])
                                        
                    if grid[check[0]][check[1]] in valid:
                        graph[loc].append( check )
            
    return graph

# Normal way to make a graph
def make_graph_not_used(grid, rows, cols):
    graph = dict()
    
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == '%':
                continue
                
            loc = (i, j)
            valid = ['-', 'P', '.']
            
            # initialize location neighbors to empty list
            graph[loc] = []
            
            if i > 0: # UP
                if grid[i - 1][j] in valid:
                    graph[loc].append( (i - 1, j) ) 
            if j > 0: # LEFT
                if grid[i][j - 1] in valid:
                    graph[loc].append( (i, j - 1) )
            if j < cols: # RIGHT
                if grid[i][j + 1] in valid:
                    graph[loc].append( (i, j + 1) )
            if i < rows: # DOWN
                if grid[i + 1][j] in valid:
                    graph[loc].append( (i + 1, j) )
                    
    return graph

# Dikstra's Algorithm
def find_food_not_used(graph, pacman, food):

    explorePath = []
    pathToFood = []
    
    visited = set()
    visited.add(pacman)
        
    # track closest predecessor from any node to pacman
    predecessors = dict()
    
    # distances from pacman to any node
    distances = dict()
    for i in range(rows):
        for j in range(cols):
            distances[(i, j)] = inf
            predecessors[(i, j)] = None
    
    distances[pacman] = 0
    
    Q = deque()
    Q.append(pacman)
    visited.add(pacman)
    
    # pop locations until the queue is empty
    while len(Q) > 0:
        
        # get the location in line
        location = Q.popleft()
        
        
        
        # track order of explored locations
        explorePath.append(location)
                    
        
#         print(location)
        
        # did we find it?
        if location == food:
            node = location
            
            # backtrack to the beginning
            while node is not None:
                grid[node[0]][node[1]] = 'x' # for testing
                pathToFood.append(node)
                node = predecessors[node]
                
            # reverse pathToFood so it is in order
            return pathToFood[::-1], explorePath
        
        # push next locations
        for node in graph[location]:
            if node not in visited:
                visited.add(node)
                explorePath.append(node)
                Q.append(node)
                
                # only update predecessor if closer to pacman than current pred...
                if distances[location] + 1 < distances[node]:  
                    predecessors[node] = location
                    distances[node] = distances[location] + 1
               
    print("Food was not found!")
    return None
    
    return explorePath, foodPath

# BFS
def find_food(graph, pacman, food):

    explorePath = []
    pathToFood = []
    
    visited = set()
    visited.add(pacman)
    
    Q = deque()
    Q.append([pacman])
    visited.add(pacman)
    
    # pop locations until the queue is empty
    while len(Q) > 0:
        
        # get the location in line
        path = Q.popleft()
                
        # track order of explored locations
#         print(f"appending: {path[-1]}")
        
#         if path[-1] in explorePath:
#             print(f"DUPLICATE! {path[-1]}")
        
        if path[-1] not in visited:
            explorePath.append(path[-1])
                            
        # did we find it?
        if path[-1] == food:
            return path, explorePath
        
        # push next locations
        for node in graph[path[-1]]:
            if node not in visited:
#                 print(f"appending {path + [node]}")
                visited.add(node)
                explorePath.append(node)
                Q.append(path + [node])

               
    print("Food was not found!")
    return None
    
    return explorePath, foodPath

def solve(grid, rows, cols, pacman, food):
    graph = make_graph(grid, rows, cols)
    pathToFood, explorePath = find_food(graph, pacman, food)
    
    print(len(explorePath) - 1)
    
#     for line in grid:
#         print("".join(line))
        
    for location in explorePath:
        print(location[0], location[1])
    
    print(len(pathToFood) - 1)
    
    for location in pathToFood:
        print(location[0], location[1])
            
solve(grid, rows, cols, pacman, food)
            
            
            
                

```
